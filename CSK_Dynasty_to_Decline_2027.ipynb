# CSK: Dynasty to Decline — A Data-Driven Prediction for 2027

## Introduction
Chennai Super Kings won IPL in 2021 and 2023.
But 2025 was a shocking decline.
This analysis uses real ball-by-ball data (2021-2026)
to find WHY CSK declined and predict 2027.

## Dataset
- Source: Kaggle IPL Dataset 2008-2026
- Rows: 283,678 ball-by-ball records
- Seasons: 2021-2026
- Tools: Python, Pandas, Matplotlib, Google Colab

```
# @title
# Match Phases Analysis - CSK Bowling
csk_bowl = df[
    (df['bowling_team'] == 'Chennai Super Kings') &
    (df['season'].astype(str).str[:4].astype(int) >= 2021)
]

def phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Death'

csk_bowl = csk_bowl.copy()
csk_bowl['phase'] = csk_bowl['over'].apply(phase)

phase_stats = csk_bowl.groupby('phase').agg(
    economy=('runs_total', lambda x: round(x.sum()/len(x)*6, 2)),
    wickets=('bowler_wicket', 'sum')
).reset_index()

print(phase_stats)

# Chart
plt.figure(figsize=(10,5))
colors = ['#FFD700', '#FF6B6B', '#4FC3F7']
plt.bar(phase_stats['phase'], phase_stats['economy'], color=colors)
plt.title('CSK Bowling Economy by Phase (2021–2026)\nWhich phase is weakest?')
plt.ylabel('Economy Rate')
plt.tight_layout()
plt.show()
```
